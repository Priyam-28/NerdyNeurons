## **Load Data**

```python
import pandas as pd
df =pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

### Data Seperation

```python
y=df['logS']
y
```

```python
x= df.drop('logS',axis=1)
x
```

### Data Splitting

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
x_test
```

## Model Buiding

### Linear Regression

**Training the model**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

**Applying the model**

```python
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

### **Evalute Model Performance**

```python
y_train
```

```python
y_lr_train_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)
lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
```

```python
lr_train_mse
```

```python
lr_train_r2
```

```python
lr_test_mse
```

```python
lr_test_r2
```

```python
lr_results=pd.DataFrame(['Linear Regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results.columns=['Model','Train MSE','Train R2','Test MSE','Test R2']
lr_results
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

```python
y_rf_train_pred=rf.predict(x_train)
y_rf_test_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
rf_train_mse=mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2=r2_score(y_train,y_rf_train_pred)
rf_test_mse=mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2=r2_score(y_test,y_rf_test_pred)
```

```python
rf_results=pd.DataFrame(['Random Forest',rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_results.columns=['Model','Train MSE','Train R2','Test MSE','Test R2']
rf_results
```

## Model Comparision

```python
df_models=pd.concat([lr_results,rf_results],ignore_index=True)
df_models
```

## Data Visualisation


```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(5,5))


z=np.polyfit(y_train,y_lr_train_pred,1)
p=np.poly1d(z)

plt.plot(y_train,p(y_train),"r")

plt.scatter(x=y_train, y=y_lr_train_pred, alpha=0.3)
plt.xlabel('Experimental logS')
plt.ylabel('Predicted logS')
plt.title('Linear Regression')
plt.show()
```
